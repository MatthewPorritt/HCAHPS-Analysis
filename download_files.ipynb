```python
import os
import sys
import shutil
import wget
import zipfile

# Make folders
work_dir = os.getcwd()
if os.path.exists("downloaded_data"):
    shutil.rmtree("downloaded_data")
if os.path.exists("work_data"):
    shutil.rmtree("work_data")
os.makedirs("downloaded_data")
os.makedirs("work_data")
```

```python
# Download all data from https://data.cms.gov/provider-data/archived-data/hospital into work directory
download_links = [
    #2017
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2017/hos_revised_flatfiles_archive_04_2017.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2017/hos_revised_flatfiles_archive_07_2017.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2017/hos_revised_flatfiles_archive_10_2017.zip",
    #2018
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2018/hos_revised_flatfiles_archive_01_2018.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2018/hos_revised_flatfiles_archive_05_2018.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2018/hos_revised_flatfiles_archive_07_2018.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2018/hos_revised_flatfiles_archive_10_2018.zip",
    #2019
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2019/hos_revised_flatfiles_archive_03_2019.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2019/hos_revised_flatfiles_archive_04_2019.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2019/hos_revised_flatfiles_archive_07_2019.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2019/hos_revised_flatfiles_archive_10_2019.zip",
    #2020
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2020/hos_revised_flatfiles_archive_01_2020.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2020/hos_revised_flatfiles_archive_04_2020.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2020/hospitals_archive_07_2020.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2020/hospitals_archive_10_2020.zip",
    #2021
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2021/hospitals_01_2021.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2021/hospitals_03_2021.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2021/hospitals_04_2021.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2021/hospitals_07_2021.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2021/hospitals_10_2021.zip",
    #2022
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2022/hospitals_01_2022.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2022/hospitals_04_2022.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2022/hospitals_07_2022.zip",
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2022/hospitals_10_2022.zip"
    #2023 to be updated
    "https://data.cms.gov/provider-data/sites/default/files/archive/Hospitals/2023/hospitals_01_2023.zip"
]

total_folders = len(download_links)
num_downloaded = 0

for url in download_links:
    num_downloaded += 1
    print("Downloading and unzipping file " + str(num_downloaded) + " of " + str(total_folders) + "...")
    # idx of char - len = negative idx of char
    slash_idx = url[::-1].index('/')
    filename_start = len(url) - slash_idx
    file_name = url[filename_start:]
    #print("file_name is " + file_name)
    downloaded_file = wget.download(url, "downloaded_data\\" + file_name)
    #downloaded_file = "downloaded_data\\" + file_name
    print("\nDownloaded " + url + " into " + downloaded_file)
    
    # Unzip and rename folders to year_month_title
    year = file_name[-8:-4]
    month = file_name[-11:-9]
    title = file_name[0:(len(file_name)-12):1]
    renamed_file = "data_" + year + "_" + month + "_" + title
    #print("renamed_file is " + renamed_file)
    extract_dir = "work_data\\" + renamed_file
    if os.path.exists(extract_dir):
        shutil.rmtree(extract_dir)
    os.makedirs(extract_dir)
    with zipfile.ZipFile(downloaded_file, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print("Extracted " + downloaded_file + " into " + extract_dir + ".\n")
print("Done.)
```
